```python
#a1

def get_common_players(list1, list2):
    common_players = []
    for player in list1:
        if player in list2:
            common_players.append(player)
    return common_players

def get_either_or_players(list1, list2):
    either_or_players = []
    for player in list1:
        
        if player not in list2:
            either_or_players.append(player)
    for player in list2:
        if player not in list1:
            either_or_players.append(player)
    return either_or_players

def get_neither_players(all_students, cricket_players, badminton_players):
    neither_players = []
    for player in all_students:
        if player not in cricket_players and player not in badminton_players:
            neither_players.append(player)
    return neither_players

def get_cricket_football_not_badminton(cricket_players, football_players, badminton_players):
    result_players = []
    for player in cricket_players:
        if player in football_players and player not in badminton_players:
            result_players.append(player)
    return result_players

# Input from the user for group A
group_a = input("Enter names of students in group A (comma-separated): ").split(',')

# Input from the user for group B
group_b = input("Enter names of students in group B (comma-separated): ").split(',')

# Input from the user for group C
group_c = input("Enter names of students in group C (comma-separated): ").split(',')

all_students = group_a + group_b + group_c

# a) List of students who play both cricket and badminton
cricket_players = group_a
badminton_players = group_b
common_players = get_common_players(cricket_players, badminton_players)
print("a) List of students who play both cricket and badminton:", common_players)

# b) List of students who play either cricket or badminton but not both
either_or_players = get_either_or_players(cricket_players, badminton_players)
print("b) List of students who play either cricket or badminton but not both:", either_or_players)

# c) Number of students who play neither cricket nor badminton
neither_players = get_neither_players(all_students, cricket_players, badminton_players)
print("c) Number of students who play neither cricket nor badminton:", len(neither_players))

# d) Number of students who play cricket and football but not badminton
football_players = group_c
result_players = get_cricket_football_not_badminton(cricket_players, football_players, badminton_players)
print("d) Number of students who play cricket and football but not badminton:", len(result_players))
```

```python
#b16


def partition(array, low, high):
    pivot = array[high]
    i = low - 1
    for j in range(low, high):
        if array[j] <= pivot:
            i = i + 1
            (array[i], array[j]) = (array[j], array[i])
    (array[i + 1], array[high]) = (array[high], array[i + 1])
    return i + 1

def quickSort(array, low, high):
    if low < high:
        pi = partition(array, low, high)
        quickSort(array, low, pi - 1)
        quickSort(array, pi + 1, high)

# Taking marks as input from the user
data = [int(x) for x in input("Enter marks separated by space: ").split()]

print("Unsorted Array")
print(data)

size = len(data)

quickSort(data, 0, size - 1)

print('Sorted Array in Ascending Order:')
print(data)
```

```python
#b11 A

def linear_search(roll_numbers, target):
    for roll_number in roll_numbers:
        if roll_number == target:
            return True
    return False

def sentinel_search(roll_numbers, target):
    roll_numbers.append(target)
    i = 0
    while roll_numbers[i] != target:
        i += 1
    roll_numbers.pop()  # Remove the sentinel
    return i < len(roll_numbers)

# User input
n = int(input("Enter the number of students: "))
roll_numbers = [int(input(f"Enter roll number {i + 1}: ")) for i in range(n)]
target_student = int(input("Enter the roll number to search: "))

# Search
linear_result = linear_search(roll_numbers, target_student)
sentinel_result = sentinel_search(roll_numbers, target_student)

# Display results
print("Roll Numbers:", roll_numbers)
print("Linear Search Result:", linear_result)
print("Sentinel Search Result:", sentinel_result)
```
